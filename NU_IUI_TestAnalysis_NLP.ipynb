# **Task 1 : Text Preprocessing**

In text preprocessing, various steps and techniques applied to raw text data before it is used for analysis or machine learning tasks. The primary purpose of text preprocessing is to clean, normalize, and transform the raw text into a format that is suitable for further NLP tasks.

Main aspects of text preprocessing are as follow:
>**Text Cleaning:**

>>**Noise Removal:** This involves removing irrelevant characters, symbols, and formatting issues from the text, such as HTML tags, special characters, and extra white spaces.

>>**Lowercasing:** Converting all text to lowercase to ensure consistency in text analysis, as "word" and "Word" should be treated as the same word.

>>**Spell Checking and Correction:** Correcting common spelling errors to improve text quality.

>**Tokenization:**
Tokenization is the process of splitting text into individual words or tokens. It's a fundamental step for most NLP tasks. For example, the sentence "I love NLP" would be tokenized into ["I", "love", "NLP"].

>**Stopword Removal:**
Stopwords are common words (e.g., "the," "and," "is") that often do not carry significant meaning and can be safely removed from text to reduce dimensionality and improve processing efficiency.

>**Stemming and Lemmatization:**
Stemming involves reducing words to their root or base form (e.g., "running" to "run"). Lemmatization, on the other hand, reduces words to their dictionary or lemma form (e.g., "better" to "good"). These techniques help in reducing inflected words to their common form.

>**Removing Special Characters and Punctuation:**
Removing non-alphanumeric characters and punctuation marks that do not carry meaningful information for many NLP tasks.

>**Normalization:**
Ensuring consistent representation of text, such as converting abbreviations ("U.S.A." to "USA"), handling contractions ("can't" to "cannot"), and converting numbers to words or vice versa.

>**Handling HTML Tags and URLs:**
Stripping HTML tags from web content and handling URLs appropriately by converting them to a common format or removing them.

>**Handling Emoji and Emoticons:**
Dealing with emojis and emoticons, either by removing them or converting them to textual representations.

>**Text Encoding:**
Converting text data into a numerical format suitable for machine learning models, often using techniques like one-hot encoding or word embeddings (e.g., Word2Vec, GloVe).

Text preprocessing plays a crucial role in NLP because the quality of the data used for analysis or modeling directly impacts the performance and accuracy of NLP applications, including text classification, sentiment analysis, machine translation, and more. By carefully preprocessing text data, NLP practitioners can ensure that the text is in a clean and standardized format, making it easier for models to learn meaningful patterns and relationships within the data.

The polarity of the sentiment (how positive or negative the text is) is extracted from the sentiment tuple. Polarity ranges from -1 (very negative) to 1 (very positive). The subjectivity of the sentiment (how subjective or objective the text is) is also extracted. Subjectivity ranges from 0 (very objective) to 1 (very subjective). # **Task 4 : Topic modelling**

Topic modelling is a technique used to discover abstract topics within a collection of text documents. The primary goal is to uncover latent topics or themes within a set of documents without requiring prior knowledge or manual labelling. It helps to organize and understand the underlying structure of a text corpus.
One popular algorithm for topic modelling is Latent Dirichlet Allocation (LDA). We can use the gensim library in Python to perform topic modelling using LDA. LDA assumes that each document is a mixture of topics, and each topic is a mixture of words. The algorithm iteratively identifies these topics and the distribution of words within them. In topic modelling, a "topic" is not a single word but rather a collection of words that are related and tend to co-occur within documents. These topics are represented as word distributions, showing the probability of each word belonging to a particular topic.

> **Applications:**

>> **Content Recommendation:** Identifying and recommending similar articles or documents based on shared topics.

>>**Information Retrieval:** Enhancing search engines by allowing users to explore content by topics.

>>**Content Summarization:** Automatically generating summaries of documents based on their dominant topics.

>>**Content Understanding:** Aiding in content analysis, categorization, and understanding trends in large text datasets.

In general topic modelling is a valuable technique in text analysis for discovering the hidden structure and thematic content within textual data. It's widely used in various fields, including academia, business, and information retrieval, to gain insights from large collections of text documents.

The code below performs topic modelling on a collection of text documents using the Gensim library, specifically the Latent Dirichlet Allocation (LDA) algorithm.

```python
#********************************
# Developer: Eaby Kollonoor Babu
# Version: 2.2
# Added: 2023-08-03
# Contact:eaby.asha@gmail.com
#********************************

import gensim
from gensim import corpora
from gensim.models import LdaModel
from gensim.utils import simple_preprocess
from nltk.corpus import stopwords

# Download NLTK resources (only needs to be done once)
import nltk
nltk.download('stopwords')

# Input text (replace this with your desired text)
input_text = """
"It's a really special part of the world," says marine biologist Taryn Foster from the Abrolhos Islands, 40 miles from the coast of Western Australia.

"There are no palm trees or luscious vegetation. But once you get in the water, you see all these tropical species of coral and fish."

Corals are animals called polyps, found mostly in tropical waters. The soft-bodied polyp forms a hard outer shell by extracting calcium
carbonate from the sea. Over time those hard shells build up to form the foundations of the reefs we see today.

Coral reefs may only cover 0.2% of the seafloor, but they provide a habitat to more than a quarter of marine species.

However, the creatures are sensitive to heat and acidification so in recent years, as the oceans have warmed and become more acidic, corals have become vulnerable to disease and death.

Damaged corals turn white, a process known as bleaching, something Ms Foster has witnessed first hand.

According to the Global Coral Reel Monitoring Network, a 1.5C increase in water temperature could see losses of between 70% and 90% of the world's reefs.

Some scientists think, that by 2070, they'll be gone altogether.
"""

# Preprocessing: Tokenize, remove stopwords, and create a dictionary
stop_words = set(stopwords.words('english'))
texts = [simple_preprocess(doc) for doc in input_text.split('\n') if doc.strip()]
dictionary = corpora.Dictionary(texts)

# Create a Bag of Words (BoW) representation
corpus = [dictionary.doc2bow(text) for text in texts]

# Build an LDA model
num_topics = 5  # Adjust the number of topics as needed
lda_model = LdaModel(corpus, id2word=dictionary, num_topics=num_topics, passes=10)

# Print the topics
for topic_id, topic in lda_model.print_topics():
    print(f"Topic {topic_id + 1}: {topic}\n")
```

# **Task 5 : Text Summarization**

Text summarization is a broad topic in Natural Language Processing (NLP), and there are various ways to approach it. A common method is to use extractive summarization, where sentences are ranked based on their importance and then the top sentences are extracted to form a summary. In other words we can say text summarization is to create a shorter, coherent, and informative summary of a longer document, such as an article, research paper, or news article. The summary should capture the essential points, main ideas, and key details while removing redundant or less relevant content. We mainly have 2 types of summarizations, one is Extractive Summarization and the ther other is Abstractive Summarization.

Extractive Summarization involes selecting and extracting sentences or phrases directly from the original text that are deemed most important. Extractive summarization relies on ranking and selecting existing content and is often based on sentence importance scores. Where as Abstractive Summarization goes beyond mere extraction and generates summaries in a more human-like way. It involves paraphrasing and rephrasing the content to create concise and coherent summaries. Abstractive summarization typically requires more advanced natural language generation techniques.



> **Applications:**

>> **Information Retrieval:** Summaries make it easier for users to quickly understand the content of long documents or web pages.

>>**News Summarization:** Automatically generating brief news summaries from longer articles or news feeds.

>>**Document Summarization:** Creating abstracts or executive summaries for research papers, reports, and legal documents.

>>**Content Generation:** Generating concise descriptions for product listings or search results.

Text summarization is a valuable tool for processing and distilling large volumes of text data, making it more digestible and accessible for readers, researchers, and applications that require concise information extraction from textual sources.

Code below performs a simple text summarization using the PageRank algorithm.

```python
#********************************
# Developer: Eaby Kollonoor Babu
# Version: 2.2
# Added: 2023-08-03
# Contact:eaby.asha@gmail.com
#********************************

import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
import networkx as nx
import nltk
from nltk.corpus import stopwords
from sklearn.metrics.pairwise import cosine_similarity

# Ensure you have the necessary resources
nltk.download('punkt')
nltk.download('stopwords')

def preprocess_sentence(sentence):
    """Preprocesses a sentence by tokenizing and removing stopwords"""
    stop_words = set(stopwords.words('english'))
    sentence = nltk.word_tokenize(sentence)
    sentence = [word for word in sentence if word not in stop_words]
    return sentence

def build_similarity_matrix(sentences):
    """Builds a similarity matrix for the sentences using TF-IDF"""
    tfidf_vectorizer = TfidfVectorizer()
    tfidf_matrix = tfidf_vectorizer.fit_transform(sentences)
    similarity_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)
    return similarity_matrix

def summarize_text(text, num_sentences=5):
    """Summarize the text based on the number of sentences given"""
    sentences = nltk.sent_tokenize(text)
    preprocessed_sentences = [preprocess_sentence(sentence) for sentence in sentences]

    similarity_matrix = build_similarity_matrix([" ".join(sentence) for sentence in preprocessed_sentences])

    # Rank sentences in similarity matrix
    sentence_similarity_graph = nx.from_numpy_array(similarity_matrix)
    scores = nx.pagerank(sentence_similarity_graph)

    # Sort the rank and pick top sentences
    ranked_sentences = sorted(((scores[i], s) for i, s in enumerate(sentences)), reverse=True)
    summarized_text = " ".join([ranked_sentence[1] for ranked_sentence in ranked_sentences[:num_sentences]])
    return summarized_text

# Your long text goes below. You can add multiple sentences or even multiple paragraphs.
text = """The use of aerial swarms to solve real-world problems has been
increasing steadily, accompanied by falling prices and improving performance
of communication, sensing, and processing hardware. The commoditization of
hardware has reduced unit costs, thereby lowering the barriers to entry to
the field of aerial swarm robotics. A key enabling technology for swarms is
the family of algorithms that allow the individual members of the swarm to
communicate and allocate tasks amongst themselves, plan their trajectories,
and coordinate their flight in such a way that the overall objectives of the
swarm are achieved efficiently. These algorithms, often organized in a
hierarchical fashion, endow the swarm with autonomy at every level,
and the role of a human operator can be reduced, in principle, to interactions
at a higher level without direct intervention. This technology depends on the
clever and innovative application of theoretical tools from control and
estimation. This paper reviews the state of the art of these theoretical
tools, specifically focusing on how they have been developed for, and applied
to, aerial swarms. Aerial swarms differ from swarms of ground-based vehicles in
two respects: they operate in a three-dimensional space and the dynamics of
individual vehicles adds an extra layer of complexity. We review dynamic modeling
and conditions for stability and controllability that are essential in order to
achieve cooperative flight and distributed sensing. The main sections of this
paper focus on major results covering trajectory generation, task allocation,
adversarial control, distributed sensing, monitoring, and mapping. Wherever
possible, we indicate how the physics and subsystem technologies of aerial
robots are brought to bear on these individual areas.
"""

print(summarize_text(text, 2))
```

**Import Libraries:** The code first imports the necessary libraries, including NumPy, scikit-learn's **TfidfVectorizer** for TF-IDF computation, NetworkX for graph-based analysis, and NLTK for natural language processing.

**Resource Download:** It downloads necessary resources from NLTK, specifically the tokenization data and stopwords for English text.

**Text Preprocessing:**

```python
#********************************
# Developer: Eaby Kollonoor Babu
# Version: 2.2
# Added: 2023-08-04
# Contact:eaby.asha@gmail.com
#********************************

import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer

def generate_text(prompt_text, model_name='gpt2-medium', max_length=150, temperature=1.0, top_k=50):
    """
    Generate text using a pretrained GPT model.

    Parameters:
    - prompt_text: initial text to start generation.
    - model_name: model size of GPT. Can be 'gpt2-small', 'gpt2-medium', 'gpt2-large', 'gpt2-xl'.
    - max_length: max number of tokens in the output text.
    - temperature: sampling softmax temperature. Higher values make the output more random.
    - top_k: keep only top k tokens with highest probability.

    Returns:
    - Generated text
    """
    # Load pre-trained model and tokenizer
    model = GPT2LMHeadModel.from_pretrained(model_name)
    tokenizer = GPT2Tokenizer.from_pretrained(model_name)

    # Convert prompt text to tensor
    input_tensor = tokenizer.encode(prompt_text, return_tensors='pt')

    # Generate text
    with torch.no_grad():
        generated = model.generate(input_tensor, max_length=max_length, temperature=temperature, top_k=top_k, pad_token_id=tokenizer.eos_token_id)

    # Convert tensor to text
    generated_text = tokenizer.decode(generated[0], skip_special_tokens=True)

    return generated_text

# You can give any input prompt and the model will generate an expanded version of it.
prompt = "The Mars One"
print(generate_text(prompt))
```
